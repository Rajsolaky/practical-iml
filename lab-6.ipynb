```python
import numpy as np
import pandas as pd
```

# 1. to convert a NumPy array to a Pandas series

```python
array = np.array([11,12,13,14,15,16,16,17,18,19,20])

print("Numpy Array :",array)

# convert Array in to padas
series = pd.Series(array)

print("padas Series :")
print(series)
```

# 2. to convert the first column of a DataFrame as a Series

```python
data = {'Column1': [10, 20, 30, 40, 50],
        'Column2': [11, 22, 33, 44, 55]}
df = pd.DataFrame(data)
# nf = pd.DataFrame(data)

#  Convert the first column of the DataFrame to a Series
first_column_series = df['Column1']
# second_column_series = nf['Column2']
print("First Column as a Pandas Series:")
print(first_column_series)
# print(second_column_series)
print()
```

# 3. to create the mean and standard deviation of the data of a given Series

```python
sample_series = pd.Series([1, 2, 3, 4, 5])
mean = sample_series.mean()
std_dev = sample_series.std()
print("Mean:", mean)
print("Standard Deviation:", std_dev)
print()
```

# 4. to sort a given Series

```python
unsorted_series = pd.Series([3, 1, 4, 2, 5])
sorted_series = unsorted_series.sort_values()
print("Sorted Series:")
print(sorted_series)
```
